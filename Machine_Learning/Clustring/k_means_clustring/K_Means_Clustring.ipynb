```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
x = dataset.iloc[:,[3,4]].values
```

```python
wcss = []
```

```python
from sklearn.cluster import KMeans
for i in range(1,11):
    kmeans = KMeans(n_clusters = i,init = "k-means++",random_state = 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("Elbow method")
plt.xlabel("Number of clustors")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5,init = "k-means++",random_state = 42)
y_pred = kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_pred == 0,0],x[y_pred == 0,1],label = "clustor 1",color = "red",s = 20)
plt.scatter(x[y_pred == 1,0],x[y_pred == 1,1],label = "clustor 2",color = "blue",s = 20)
plt.scatter(x[y_pred == 2,0],x[y_pred == 2,1],label = "clustor 3",color = "pink",s = 20)
plt.scatter(x[y_pred == 3,0],x[y_pred == 3,1],label = "clustor 4",color = "yellow",s = 20)
plt.scatter(x[y_pred == 4,0],x[y_pred == 4,1],label = "clustor 5",color = "green",s = 20)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],label = "centroid ",color = "black",s = 100)
plt.title("k means clustring")
plt.xlabel("income")
plt.ylabel("spending score")
plt.legend()
plt.show()
```
