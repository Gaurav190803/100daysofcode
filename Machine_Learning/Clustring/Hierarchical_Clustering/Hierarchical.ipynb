```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
x = dataset.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
dendogrram = sch.dendrogram(sch.linkage(x, method = "ward"))
plt.title("dendogram")
plt.xlabel("Customers")
plt.ylabel("euclid distance")
plt.show()

```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = "euclidean", linkage = "ward")
y_pred = hc.fit_predict(x)
```

```python
y_pred
```

```python
plt.scatter(x[y_pred == 0,0],x[y_pred == 0,1],label = "clustor 1",color = "red",s = 20)
plt.scatter(x[y_pred == 1,0],x[y_pred == 1,1],label = "clustor 2",color = "blue",s = 20)
plt.scatter(x[y_pred == 2,0],x[y_pred == 2,1],label = "clustor 3",color = "pink",s = 20)
plt.scatter(x[y_pred == 3,0],x[y_pred == 3,1],label = "clustor 4",color = "yellow",s = 20)
plt.scatter(x[y_pred == 4,0],x[y_pred == 4,1],label = "clustor 5",color = "green",s = 20)

plt.title("k means clustring")
plt.xlabel("income")
plt.ylabel("spending score")
plt.legend()
plt.show()
```
